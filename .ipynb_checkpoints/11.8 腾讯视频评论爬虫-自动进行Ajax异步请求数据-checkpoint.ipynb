# 自动进行Ajax异步请求数据

```python
#1·腾讯视频评论爬虫（单页）http://localhost:8888/notebooks/notes/11.8%20%E8%85%BE%E8%AE%AF%E8%A7%86%E9%A2%91%E8%AF%84%E8%AE%BA%E7%88%AC%E8%99%AB.ipynb#自动进行Ajax异步请求数据
from urllib import request
import re 
#https://video.coral.qq.com/filmreviewr/c/upcomment/{视频ID}?commentid={评论ID}&reqnum={每次提取的评论个数}
vid='i5w51tl7vbl5mid'
cid='6320117440705909416'
num='20'
#构造当前评论网址
url='https://video.coral.qq.com/filmreviewr/c/upcomment/'+vid+'?commentid='+cid+'&reqnum='+num
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
opener=request.build_opener()
headall=[]
for key,value in headers.items():
    item=(key,value)
    headall.append(item)
opener.add_headers=headall
request.install_opener(opener)
#爬取网页
data=request.urlopen(url).read().decode('utf-8')
titlepat='"title":"(.*?)"'
commentpat='"content":"(.*?)"'
titleall=re.compile(titlepat,re.S).findall(data)
commentall=re.compile(commentpat,re.S).findall(data)
for i in range(0,len(commentall)):
    try:
        print('评论标题是：'+eval('u"'+titleall[i]+'"'))
        print('评论内容是：'+eval('u"'+commentall[i]+'"'))
        print('--------')
    except Exception as err:
        print(err)
```

```python
#2·腾讯视频评论爬虫（自动切换下一页）
from urllib import request
import re 
#https://video.coral.qq.com/filmreviewr/c/upcomment/{视频ID}?commentid={评论ID}&reqnum={每次提取的评论个数}
vid='i5w51tl7vbl5mid'
cid='6320117440705909416'
num='20'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
opener=request.build_opener()
headall=[]
for key,value in headers.items():
    item=(key,value)
    headall.append(item)
opener.add_headers=headall
request.install_opener(opener)
for j in range(0,10):
    #爬取网页
    print('第'+str(j+1)+'页')
    thisurl='https://video.coral.qq.com/filmreviewr/c/upcomment/'+vid+'?commentid='+cid+'&reqnum='+num
    data=request.urlopen(thisurl).read().decode('utf-8')
    titlepat='"title":"(.*?)","abstract":"'
    commentpat='"content":"(.*?)"'
    titleall=re.compile(titlepat,re.S).findall(data)
    commentall=re.compile(commentpat,re.S).findall(data)
    lastpat='"last":"(.*?)"'
    cid=str(re.search(lastpat,data,re.S))
    for i in range(0,len(commentall)):
        try:
            print('评论标题是：'+eval('u"'+titleall[i]+'"'))
            #commentall[i]=commentall[i].replace('"','')
            print('评论内容是：'+eval('u"'+commentall[i]+'"'))
            print('--------')
        except Exception as err:
            print(err)
```
